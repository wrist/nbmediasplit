```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import Audio, Image, display
```

```python
fs = 48000.0
f = 1000.0
ts = np.arange(1.0, 10.0 * fs)
ys = np.sin(ts * 2.0 * np.pi * f / fs)
```

```python
# single figure
fig = plt.figure(1)
ax = fig.add_subplot(1,1,1)
end_pt = int(0.1*fs)
ax.plot(ts[:end_pt] / fs, ys[:end_pt], label="sin wave {}[Hz]".format(f))
ax.legend(loc=1)
ax.set_xlabel("time [s]")
ax.set_ylabel("amplitude")
ax.grid(b=True)

fig.savefig("test1.png")
```

```python
# subplot
fig = plt.figure(2)
ax = fig.add_subplot(2,1,1)
end_pt = int(0.1*fs)
ax.plot(ts[:end_pt] / fs, ys[:end_pt], label="sin wave {}[Hz]".format(f))
ax.legend(loc=1)
ax.set_xlabel("time [s]")
ax.set_ylabel("amplitude")
ax.grid(b=True)

ax = fig.add_subplot(2,1,2)
ax.plot(ts / fs, ys, label="sin wave {}[Hz]".format(f))
ax.legend(loc=1)
ax.set_xlabel("time [s]")
ax.set_ylabel("amplitude")
ax.grid(b=True)

fig.savefig("test2.png")
```
